```
import random
import numpy as np
from random import randint
```

```
# Variables
n = 8
mutationRate = 0.2  
totalPopulation = 10  
generations = 1000
 
#Initial list of alpha elements
alpha_list = [x for x in range(0, n)] 
alpha_list
```

```
 #Function Defnitions
 
#used for toss
def random_toss():
    return random.choice([0,1])
 
#get random character
def get_randomdata():
    return random.choice(alpha_list)

#Create a random new chess set position
def new_dataset():
    return [get_randomdata() for x in range(n)]


#Moves the columns in chess board
def new_roll(base,move_by):
    if(move_by > 0):
        return np.append(np.zeros(move_by),base[0:n-move_by])
    if(move_by < 0):
        return np.append(base[-move_by:],np.zeros(-move_by))

 #calculate fitness score for one chess set entity

def calculate_fitness(child):
    
    #define variables
    score     = 0
    max_score = 0
    cube      = []
    
    #configure the board
    for i in range(n):
        base = np.zeros(n, dtype = int)
        base[child[i]] =1 
        base = base[::-1]
        cube.append(base)

    #sideways fitness    
    score += len(np.unique(child))
    max_score += n 

    #crossways fitness    
    for i in range(n):
        for j in range(i+1,n):
            threat = 0

            roll1 = new_roll(cube[j],j-i)   #right position shift
            roll2 = new_roll(cube[j],i-j)   #left position shift
            
            if sum((cube[i] == roll2)*1) == 8:   
                threat = 1                     
            if sum((cube[i] == roll1)*1) == 8:    
                threat = 1                     
                 
            if threat ==0:                    #if there is no threat to queen promote scores
                score += 1                    
            max_score += 1                       
    
    return round(score/max_score,3)  # returns the percentage of fitness

  #example
a = [0,1,2,3,4,5,6,7]
print("Fitness of ",a, "is:" , str(measure_fitness(a)) )
```

```
#crossover function to create new

def generate(parent1,parent2):    
    #Crossover
    child = []
    for i in range(n):        
        toss = random_toss()
        
        if toss== 1:
            child.append(parent1[i])
        else:
            child.append(parent2[i])
    
    #Mutation
    for a in range(int(np.ceil(n* mutationRate))):
        child[get_randomdata()] = get_randomdata()         

    return child

# helps sorts the fitness function
def srt_fitness(x):
    return x[1]

#Fittest elements
def survival():
    global population
    
    for i in range(len(population)):
        element = population[i][0]
        fitness = calculate_fitness(element)
        population[i][1] = fitness
    
    #sort in order of fitness
    population.sort(key = srt_fitness, reverse = True)
    
     #Survival of the fittest
    population = population[:totalPopulation]
```

```
#MAIN CODE
#STEP 0: 
#Initial population
population = []
for i in range(totalPopulation):
    population.append([new_dataset(),0])
    
# The Generation Loop
for i in range(1,generations):

    #STEP 1:
    #Generate population
    for m in range(totalPopulation):
        for j in range(totalPopulation):

            parent1 = population[m][0]
            parent2 = population[j][0]
            child = generate(parent1,parent2)
            
            #add to general population
            population.append([child,0])   #initial fitness value = 0

    #STEP 2:
    #Survival using fitness score
    survival()
        
    #Print Generations
    print("Generation No.: ", i)
    print("Fitness score :", population[0][1], population[0][0])
    print()
    
    # End if sequence generated
    if population[0][1] == 1:
        break
 
```
